# Applying basic column transformation

```python
#Replacing 2 with 0 for chronic conditions, 0 indicates no chronic condition , 1 -> yes for chronic conditions
df=df.replace({'ChronicCond_Alzheimer': 2, 'ChronicCond_Heartfailure': 2, 'ChronicCond_KidneyDisease': 2,
                           'ChronicCond_Cancer': 2, 'ChronicCond_ObstrPulmonary': 2, 'ChronicCond_Depression': 2,
                           'ChronicCond_Diabetes': 2, 'ChronicCond_IschemicHeart': 2, 'ChronicCond_Osteoporasis': 2,
                           'ChronicCond_rheumatoidarthritis': 2, 'ChronicCond_stroke': 2 }, 0)
```

```python
df[['ChronicCond_Alzheimer','ChronicCond_Heartfailure']]
```

```python
#0->male,1->female
df=df.replace({'Gender':2},0)
```

```python
#0->no , 1-> yes
df=df.replace({'RenalDiseaseIndicator':'Y'},1)
```

```python
# Replace 'Yes'/'No' with 1/0 and ensure integer type
# Replace 'Yes'/'No' with 1/0
df['PotentialFraud'] = df['PotentialFraud'].map({'Yes': 1, 'No': 0})

# Fill missing values (if any) with 0 or your desired default
df['PotentialFraud'] = df['PotentialFraud'].fillna(0).astype(int)
```
